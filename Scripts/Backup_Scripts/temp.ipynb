```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (make sure the path is correct)
df = pd.read_csv("weather_crime_final.csv")

# Aggregate total crime by city
crime_by_city = df.groupby("city")["total_crime"].sum()

# Print values (helpful for copy-paste into appendix)
print("\nTotal Crime by City:")
print(crime_by_city)

# Create pie chart
plt.figure()
plt.pie(crime_by_city, labels=crime_by_city.index, autopct='%1.1f%%')
plt.title("Share of Total Crime by City")

plt.show()
```
