```python
import requests
import pandas as pd
import time

url = "https://data.cityofchicago.org/resource/ijzp-q8t2.json"

limit = 50000
offset = 0
all_rows = []

while True:
    params = {
        "$select": "id,date,primary_type,arrest,domestic,latitude,longitude",
        "$where": "year between 2020 and 2023",
        "$limit": limit,
        "$offset": offset,
        "$order": "date"
    }

    response = requests.get(url, params=params)
    response.raise_for_status()

    batch = response.json()
    print(f"Downloaded {len(batch)} rows at offset {offset}")

    if len(batch) == 0:
        break

    all_rows.extend(batch)
    offset += limit

    time.sleep(0.2)  # prevents API rate limiting

df = pd.DataFrame(all_rows)

# convert timestamp
df["date"] = pd.to_datetime(df["date"])

# extract hour
df["hour"] = df["date"].dt.hour

# save file
df.to_csv("chicago_crime_2020_2023.csv", index=False)

print(df.head())
print("Total rows:", len(df))
```
